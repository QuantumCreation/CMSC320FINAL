<h1> How Do You Measure Human Progress?</h1>

If we want to determine if the world has become a better place through data, we first need to figure out what we need to look for. I have decided a 6 different values, some of them are index's representing a number of different measures, while others are independent measurements. For example, the Economic Freedom Summary Index is created based on 12 different measures of economic freedom, whereas the Homocide rate is a single measurement. Below, I have listed each of the values I will be using for this analysis, along with what they represent.

<h2>List of Values Used to Judge Human Progress</h2>

<ul>
    <li>Development Index</li>
    <li>Homocide Rate</li>
    <li>Hunger Index</li>
    <li>Cantril Happiness Scale</li>
    <li>Gender Pay Gap Comparison</li>
    <li>Economic Freedom Index</li>
</ul> 

<h1>Development Index</h1>

The Development Index uses life expectancy, access to education, and standard of living to generate an index value representing the total development of the culture, government, and average life of a citizen.

<h1>Homocide Rate</h1>

The Homocide Rate is the number of homocides per 100,000 individuals in the country. It is a rate, not a number, hence it scales with the size of the country in question.

<h1>Hunger Index</h1>

The Hunger Index measures how well the caloric and nutritional requirements are met in a specific country to generate a value. The difference between nutrition and caloric requirements are an important thing to note. An individual could have more than enough calories to survive, but lack necessary vitamins and minerals leading to sickness, disease, or even death in extreme cases.

<h1>Cantril Happiness Scale</h1>

The Cantril Happiness data is a scale where respondents are asked to rate their lives on a scale from 1-10, where 10 is the happiest, and 1 is the most depressed.

<h1>Gender Pay Gap Comparison</h1>

The Gender Pay Gap Comparison (Equality measure) is the percent difference between the wages of men and women in various countries over time.

<h1>Economic Freedom Index</h1>

The Economic Data has many different measures, but what I will be looking at is the Economic Freedom Summary Index, a measure of how free individuals are in a country to pursue economic oppurtunities. This considers 12 different measures, which are then broken into four categories. They are:

<ul>
    <li>Rule of Law (property rights, government integrity, judicial effectiveness)</li>
    <li>Government Size (government spending, tax burden, fiscal health)</li>
    <li>Regulatory Efficiency (business freedom, labor freedom, monetary freedom)</li>
    <li>Open Markets (trade freedom, investment freedom, financial freedom)</li>
</ul> 


Again, these are in no way 100% comprehensive, but they do provide a very good clue as to whether or not the world at large is improving. Works such as Steven Pinker's "Enlightenment Now" cover a bit more ground in terms of the breadth of statistics and measures considered, though it is not quite as heavy in the data analysis itself. <h1>DATA MANAGEMENT AND CURATION</h1>

Wonerful. We have imported all of our data and we can get straight into the analysis... Or not. Look at the last set of data we have just imported. The first three rows are entirely "NaN", and the first column is also "NaN". Not only that, but we don't have the correct column names either. It looks like our column names are actually in the fourth row (index 3) in our data. This requires us to clean up and fix the data for our use. Often when we are collecting data, we will find that values are missing, the data format isn't what we need, or there are other problems. This is why we need to clean up our data to use it. That's what we will get started on below.

```python
#gets a list of the correct column names
column_names = econ_data.loc[3]
#drops the first column
econ_data = econ_data.drop(econ_data.columns[0],  axis = 1)
#drops the first four rows
econ_data = econ_data.drop([0,1,2,3])
#renames the columns with the correct column names
econ_data = econ_data.rename(columns = column_names)
econ_data
```

Great! Now we have cleaned up our data... sort of. If we take a look at the table we printed out, there are still a number of spaces filled with "NaN". There are three ways we could deal with this. We could:

<ul>
    <li>Remove rows and columns with missing data</li>
    <li>Encode missing data</li>
    <li>Impute missing data</li>
</ul>

Each of these has some pros and cons. For our case, we will be using the first option: Remove the missing data. Fortunately, the Economic Freedom Summary Index acts as a summary statistic for all of the other measures in our table. This means that the organization collecting the data has already done some of their own imputation to create the index when certain measures in the index are missing, such as the case in in row 4052 seen above where tax compliance and Business regulations have missing values. There will still be missing data when even the index is missing, but it is an improvement nonetheless.

```python
#Selects only the columns of the table we will be using
econ_data = econ_data[['Year', 'ISO_Code', 'Countries', "Economic Freedom Summary Index"]].dropna()
#This data set seems to have stored the year as a string. This corrects it
econ_data['Year'] = econ_data['Year'].astype(int)
#Again, I have corrected to data type to float
econ_data['Economic Freedom Summary Index'] = econ_data['Economic Freedom Summary Index'].astype(float)
econ_data
```

Now we have loaded all of our data and can start doing exploratory data analysis. We can't make any huge assumptions right at the beginning, so it is usually a good idea to start with some general plot of all of our values. If the plot ends up looking messy, or no meaningful conclusions can be drawn, then we can try a new method and experiment to find what we are looking for (assuming a correlation exists). If we do notice any patterns or interesting indications on our plots, we can focus on those sections and dive further in with our data analysis.

```python
#Selects only the data from 1992
hunger_data_1992 = hunger_data.loc[hunger_data['Year']==1992]
```

```python
#groups the data based on the hunger index ranges
tmp = hunger_data_1992.groupby(pd.cut(hunger_data_1992["Hunger Index"], [0, 10, 20, 35, 50, 100])).count()
hunger_data_grouped = pd.DataFrame(index = ['Low', 'Moderate', 'Serious', 'Alarming', 'Extremely Alarming'], columns = ['Nations Count'])
```

```python
#Renames the column to nations count and prints a table
hunger_data_grouped['Nations Count'] = list(tmp['Hunger Index'])
hunger_data_grouped
```

The table above shows us the number of nations that fall within the specified range.

```python
#Creates table of the hunger percentages for each category
hunger_percentages = 100*(hunger_data_grouped['Nations Count']/sum(hunger_data_grouped['Nations Count']))
hunger_percentages = pd.DataFrame(hunger_percentages)
hunger_percentages.rename(columns = {"Nations Count":"Percentage"})
```

```python
#Plots the pie chart with the given categories
plt.figure(figsize=(10, 10))
plt.title('Global Hunger Index 1992', fontsize=20)
plt.pie(hunger_data_grouped['Nations Count'], explode =(0.05, 0.05, 0.05, 0.05, 0.05), 
        labels=['Low', 'Moderate', 'Serious', 'Alarming', 'Extremely Alarming'], 
        autopct='%1.1f%%', colors = ['red', 'blue','green','yellow','purple'])
plt.show()
```

```python
#Selects only the data from 2018
hunger_data_2018 = hunger_data.loc[hunger_data['Year']==2018]
```

Next, we will take a look at the Cantril Happiness data. Looking at our previous scatter plot, we can take a guess that a violin plot would work well for the data distribution we had there. Lets get into it.

```python
#creates list of unique years in the data
years = happiness_data['Year'].unique()

#creaets an empty array to organize data by year
happiness = []

#goes year by year and places a list for a each individual year into the happiness list
for year in years:
    happiness.append(happiness_data.loc[happiness_data['Year'] == year]['Cantril Happiness'])
```

```python
plt.violinplot(happiness, years, widths=1,showmeans=True)
plt.xlabel("Year")
plt.ylabel("Cantril Happiness")
plt.title("Cantril Happiness vs Year")
```

Again, lets analyze this plot to see what we can find. We notice a couple things right at the beginning. One, the first set of data seems to be an outlier. This is likely due to which nations where given the survey during that first year. By 2006, we see a drastic change in the data compared to the previous year, which the average happiness dropping significantly. From 2006 to 2018, we see something interesting. The average happiness seems to increase just a little bit, but the shape of our distributions change more drastically. Near 2006 and 2007 we see that the distibution seems to clump near the bottom of the scale, and the distribution in 2017 and 2018 seems to be more centered. Consider ways we could improve the information we get from this plot. First we could try dropping the first year of measurement, as it seems to be a notable outlier. Then we could fit the data to a linear regression and create a residual plot. Depending on the shape of the residual plot, it could indicate the presence of an interaction between variables. Our data only has the country and the year, so we are limited in what we could test here, but in the future we could look for a different data set in order to do an analysis on the interaction. If the reader would like to look into this possible analysis more, I'd recommend the following link:

https://www.jmp.com/en_us/statistics-knowledge-portal/what-is-regression/simple-linear-regression-assumptions.html

This site helps explain how we can use residuals to analyze the effectivenessa nd validity of a linear regression. I will give an example later on, once we have created a unique plot for each of our data sets first.

```python
#defines the characteristics of the lines
plt.rc('lines', linewidth=1)
plt.rc('axes', prop_cycle=default_cycler)

#mx is used to find the maximum y value
mx = 0
#slopes is an array of the polyfit slopes for the regression lines
slopes = []

#iterates through each country and plots the Gender Wage Gap vs Year
for Country in equality_data.Entity.unique():
    df = equality_data[equality_data['Entity'] == Country]

    x = df['Year']
    y = df['Gender Wage Gap']
    
    m, b = np.polyfit(x, y, 1)
    slopes.append((Country,m))
    
    if max(y) > mx:
        mx = max(y)
    
    plt.plot(x, y, marker = '', label = Country)

#Titles and labels graphs
plt.title("Gender Wage Gap vs Year")
plt.xlabel('Year')
plt.ylabel('Gender Wage Gap (% Difference)')
plt.yticks(list(np.linspace(0, mx,10)))
plt.show()
```

```python
#Creates a table of the each country in the dataset and includes the slope for the wage gap vs year
df = pd.DataFrame(slopes, columns = ['Country', 'Yearly Percent Decrease in Gender Wage Gap']).sort_values('Yearly Percent Decrease in Gender Wage Gap')
df
```

Taking a look at this line graph, we see that in general, the gender wage gap has been decreasing over time. Only 6 countries out of 38 had an average increase in the gender wage gap, while 32 nations had a decrease in the gender wage gap. This indicates a reasonable amount of improvement over time, and supports the idea that at least across countries measured, the gender wage gap issue has been improving over time.

Wow! This plot looks nothing like our previous violin plot of the Cantril Happiness Scale. We have a huge skew towards the high end, and massive clumping near the bottom. While in many cases we can try a linear regression, in this case we can say it would definitely be a bad choice. Why? It is because when making a linear regression, we make certain assumptions about the data. We assume:

<ul>
    <li>Linearity: The relationship between X and the mean of Y is linear.</li>
    <li>Homoscedasticity: The variance of residual is the same for any value of X.</li>
    <li>Independence: Observations are independent of each other.</li>
    <li>Normality: For any fixed value of X, Y is normally distributed.</li>
</ul>

In this case, we clearly don't have normality, a requirement for the lineear regression to be accurate. It is very difficult to determine if there has been any improvement based on this plot. Lets come back to this issue after we've completed our final plot and see if there are any tools or methods we could use in this scenario.

<h1>HYPOTHESIS TESTING AND MACHINE LEARNING</h1>

We've gotten a good look at our data and now we have a good idea what we are looking for. At this point, we usually want to propose a hypothesis to test. In Hypothesis testing, we essentially have two possibilites, the Null Hypothesis and an Alternative Hypothesis. The Null Hypothesis proposes that there is no correlation between the variables. For our case, it would mean that there has been no improvement for the average human on Earth for the given time period or that any improvement that was measured is due to statistical variation and randomness. The Alternative hypothesis suggests that there is a meaningful and statistically significant correlation between the independent ande dependent variables. In our case, this means that the world has been improving meaningfully over the years in question. By performing a number of statistical methods, we can analyze with mathematical certainty whether or not the change of our vairous measures is statistically significant, or if it was due to sampling errors and randomness.

For the example in this tutorial, we will take a look at the Economic Freedom Summary Index. It is recommended that the reader try out hypothesis testing with the other data-sets to evaluate whether they agree with this example.

```python
#module to help us create the linear model
from sklearn import linear_model
```

```python
# A linear relationship between `x` and `y`.
ind = [[i] for i in econ_data['Year']]

dep = [[i] for i in econ_data['Economic Freedom Summary Index']]
```

```python
lm = linear_model.LinearRegression()
lm.fit(ind, dep)
```

```python
fig, ax = plt.subplots()
    
ax.scatter(ind, dep, marker = '.')
ax.set_xlabel("Independent")
ax.set_ylabel("Dependent")
ax.set_title("A Linear Regression")
    
ax.plot(ind, lm.predict(ind), color='red')
```

Here, we have fit our model with scikit learn. Using the funciton lm.predict, we can predict what the value of the Economic Freedom Summary Index will be for any specified year based on the linear regression. However, lets check to see if our residuals are normal as well.

```python
#Adds a column to the econ data with the predicted value
econ_data['Prediction'] = lm.predict(ind)
```

```python
#Calculates the residuals between the prediction and the actual measured values
econ_data['residuals'] = econ_data['Economic Freedom Summary Index'] - econ_data['Prediction']
years = econ_data.Year.unique()
residuals = []

#creates a sublist with all of the residuals for the specified year
for year in years:
     residuals.append(econ_data.loc[econ_data['Year'] == year]['residuals'])
    
```

```python
fig, ax = plt.subplots()
    
ax.violinplot(residuals, years, widths=2,showmeans=True)
ax.set_xlabel("Year")
ax.set_ylabel("Economic Freedom Index Residual")
ax.set_title("Freedom Index Residual vs Year")
```

As we can see, there are two different sections on this graph. From 1970 to 2000, there is around a 5 year gap between each measurement. After 2000, they seem to have measured the data every single year. This makes the data from 2000 to 2010 very clustered and difficult to read. I will create a separate violin plot to show just the range from 2000-2018.

```python
#Calculates the residuals between the prediction and the actual measured values
econ_data['residuals'] = econ_data['Economic Freedom Summary Index'] - econ_data['Prediction']
years = list(range(2000, 2018))
residuals = []

#creates a sublist with all of the residuals for the specified year
for year in years:
     residuals.append(econ_data.loc[econ_data['Year'] == year]['residuals'])
```

```python
fig, ax = plt.subplots()
    
ax.violinplot(residuals, years, widths=1,showmeans=True)
ax.set_xlabel("Year")
ax.set_ylabel("Economic Freedom Index Residual")
ax.set_title("Freedom Index Residual vs Year")
```

This residual plot shows us that from the period 2000-2018, the residuals from our linear regression are all roughly normalized which is very good, as that is one of the four assumptions for a linear regression. Lets now check that our regression for this range is statistically significant.

```python
#import module for testing our p-values
import statsmodels.api as sm
```

```python
#Creates a summary of statistical data for our prediction and original data
#We will only be looking at the coefficients and the p values here (P> |t|)
X2 = sm.add_constant(ind)
est = sm.OLS(dep, X2)
est2 = est.fit()
print(est2.summary())
```

Looking at our output, we see that both our constant and our slope both have p values much less than 0.05, indicating that they are in fact statistically significant. If we had a p value greater than 0.05, it would indicate that there was not a significant different between the data and the Null Hypothesis, ie. no correlation. From this, we can confidently say that the Economic Freedom Summary Index has been increasing of the couple decades.
